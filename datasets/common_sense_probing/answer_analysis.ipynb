```python
import json
from tqdm import tqdm

# Load the data
with open('/home/as5606/Datasets/Cholec_text_grounded_questions/hulumed_14B_grounded_accuracy_comparison.json', 'r') as f:
    data = json.load(f)

total_matches = 0
total_no_matches = 0
no_match_items = []

for item in tqdm(data):
    grounded_qa_list = item.get('grounded_qa_with_matches', [])
    
    for qa in grounded_qa_list:
        if qa['match'] == 'MATCH':
            total_matches += 1
        else:
            total_no_matches += 1
            # Collect NO_MATCH items with context
            no_match_items.append({
                'id': item['id'],
                'image': item['image'],
                'question': qa['question'],
                'original_answer': qa['original_answer'],
                'hulu_med_answer': qa['hulu_med_answer']
            })

total_qa = total_matches + total_no_matches
accuracy = (total_matches / total_qa * 100) if total_qa > 0 else 0

print(f"Total grounded QA pairs: {total_qa}")
print(f"Matches: {total_matches}")
print(f"No Matches: {total_no_matches}")
print(f"Accuracy: {accuracy:.2f}%\n")

# Show NO_MATCH examples
print(f"{'='*80}")
print(f"First 10 NO_MATCH examples:")
print(f"{'='*80}\n")
```

```python
import json
from tqdm import tqdm

# Load the data
with open('/home/as5606/Datasets/Cholec_text_grounded_questions/hulumed_14B_grounded_accuracy_comparison.json', 'r') as f:
    data = json.load(f)

# Initialize counters for each category
categories = {
    'color': {'matches': 0, 'total': 0, 'no_matches': []},
    'side': {'matches': 0, 'total': 0, 'no_matches': []},
    'tip_direction': {'matches': 0, 'total': 0, 'no_matches': []},
    'count': {'matches': 0, 'total': 0, 'no_matches': []}
}

category_indices = {
    'color': 0,
    'side': 1,
    'tip_direction': 2,
    'count': 3
}

for item in tqdm(data):
    grounded_qa_list = item.get('grounded_qa_with_matches', [])
    
    # Check each category
    for category_name, index in category_indices.items():
        if index < len(grounded_qa_list):
            qa = grounded_qa_list[index]
            categories[category_name]['total'] += 1
            
            if qa['match'] == 'MATCH':
                categories[category_name]['matches'] += 1
            else:
                categories[category_name]['no_matches'].append({
                    'id': item['id'],
                    'question': qa['question'],
                    'original_answer': qa['original_answer'],
                    'hulu_med_answer': qa['hulu_med_answer']
                })

# Print results
print(f"{'='*70}")
print(f"GROUNDED QUESTIONS ACCURACY BY CATEGORY")
print(f"{'='*70}\n")

for category_name, index in category_indices.items():
    cat = categories[category_name]
    accuracy = (cat['matches'] / cat['total'] * 100) if cat['total'] > 0 else 0
    
    print(f"{category_name.upper().replace('_', ' ')}:")
    print(f"  Total: {cat['total']}")
    print(f"  Correct (MATCH): {cat['matches']}")
    print(f"  Incorrect (NO_MATCH): {cat['total'] - cat['matches']}")
    print(f"  Accuracy: {accuracy:.2f}%")
    print()


```
