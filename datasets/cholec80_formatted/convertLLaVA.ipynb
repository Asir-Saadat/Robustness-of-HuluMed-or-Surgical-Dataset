```python
import os
import json

sentence_folder = "/home/as5606/Datasets/Cholec80-VQA/Sentence"
frames_base = "/home/as5606/Datasets/chole80_framed/cholec80/frames"
output_path = "/home/as5606/Datasets/cholec_formattted_data/cholec80_llava.json"

dataset = []

for video_folder in sorted(os.listdir(sentence_folder), key=lambda x: int(x)):
    video_path = os.path.join(sentence_folder, video_folder)
    if not os.path.isdir(video_path):
        continue
    video_name = f"video{int(video_folder):02d}"

    for qa_file in sorted(os.listdir(video_path), key=lambda x: int(x.replace("_QA.txt", ""))):
        qa_frame_num = int(qa_file.replace("_QA.txt", ""))
        png_frame_num = qa_frame_num // 25 + 1
        png_path = os.path.join(frames_base, video_name, f"{video_name}_{png_frame_num:06d}.png")

        with open(os.path.join(video_path, qa_file), "r") as f:
            for line in f:
                line = line.strip()
                if not line:
                    continue
                question, answer = line.split("|")
                entry = {
                    "id": f"{video_name}_frame{qa_frame_num:06d}",
                    "image": png_path,
                    "conversations": [
                        {"from": "human", "value": f"{question}\n<image>"},
                        {"from": "gpt",   "value": answer}
                    ]
                }
                dataset.append(entry)

with open(output_path, "w") as f:
    json.dump(dataset, f, indent=2)

print(f"Done! {len(dataset)} entries written to {output_path}")
```
