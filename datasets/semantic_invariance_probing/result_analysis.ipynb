```python
import json
from tqdm import tqdm

# Load the data
with open('/home/as5606/Datasets/Cholec_text_semantic_similar_questions/hulu_med_14B_semantic_variations_accuracy.json', 'r') as f:
    data = json.load(f)


total_correct_count = 0



for item in tqdm(data):
    item_id = item['id']
    original_answer = item['original_answer'].lower()

    correct_count = 0
    incorrect_count = 0
    
    for variation in item.get('semantic_variations_with_matches', []):

        question = variation['question'].lower()
        hulu_answer = variation['hulu_med_answer'].lower()
        match = variation['match']
        
        is_correct = False
        
        # For "how many" questions - use match field
        if "how many" in question:
            if match == 'MATCH':
                is_correct = True
        
        # For "is used" / "is not used" questions - check answer content
        elif "used" in question or "used" in original_answer:
            if "is not used" in original_answer:
                # Should say "no" or have "not"
                if "no" in hulu_answer or "not" in hulu_answer:
                    is_correct = True
            elif "is used" in original_answer:
                # Should say "yes" or have "is used"
                if "yes" in hulu_answer or "is used" in hulu_answer:
                    is_correct = True
        
        if is_correct:
            correct_count += 1
        else:
            incorrect_count += 1


    total_correct_count += correct_count/5

print(total_correct_count/len(data))
```

```python
import json
from tqdm import tqdm

with open('/home/as5606/Datasets/Cholec_text_semantic_similar_questions/hulu_med_4B_semantic_variations_accuracy.json', 'r') as f:
    data = json.load(f)

total_variance = 0

for item in tqdm(data):
    # Collect all answers for this item's variations
    answers = []
    
    for variation in item.get('semantic_variations_with_matches', []):
        hulu_answer = variation['hulu_med_answer'].lower().strip()
        answers.append(hulu_answer)
    
    # Count distinct answers
    distinct_answers = len(set(answers))
    total_variance += distinct_answers

# Calculate AVS
avs = total_variance / len(data)
print(f"Answer Variance Score (AVS): {avs:.4f}")
print(f"Interpretation:")
print(f"  - AVS = 1: Perfect consistency (same answer for all rephrasings)")
print(f"  - AVS > 1: Instability (different answers for same question)")
```
