```python
from google import genai
from google.genai import types

# Initialize client
client = genai.Client(

    api_key="os.environ.get("GOOGLE_API_KEY")"

)

# Image path and question
image_path = '/home/as5606/Datasets/chole80_framed/cholec80/frames/video31/video31_000049.png'  # Change this to your image path
question = "Do you know what the patient age is here? Answer simply"

# Read image
with open(image_path, 'rb') as f:
    image_bytes = f.read()

# Determine mime type from file extension
image_ext = image_path.lower().split('.')[-1]
mime_types = {
    'jpg': 'image/jpeg',
    'jpeg': 'image/jpeg',
    'png': 'image/png',
    'gif': 'image/gif',
    'webp': 'image/webp'
}
mime_type = mime_types.get(image_ext, 'image/jpeg')

# Generate VQA response
response = client.models.generate_content(
    model='gemini-3.1-flash-lite',
    contents=[
        types.Part.from_bytes(
            data=image_bytes,
            mime_type=mime_type,
        ),
        question
    ]
)

print(f"Question: {question}")
print(f"Answer: {response.text}")
```
